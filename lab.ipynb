4)Define a class MatrixDotProduct,implement a method to perform xtranspose * y.Define a main and pass the matrices x and y and compute x transpose * y

```python
import numpy as np
class MatrixDotProduct:
  def __init__(self,X,Y):
    self.X=X
    self.Y=Y
  def transpose_dot_product(self):
    self.X=np.transpose(self.X)
    prod=np.dot(self.X,self.Y)
    return prod
def main():
  x=np.array([[1,2,3],[5,8,9]])
  y=np.array([[7,8,9],[10,11,12]])
  obj=MatrixDotProduct(x,y)
  prod=obj.transpose_dot_product()
  print(prod)
if __name__=="__main__":
  main()
```

Read two vectors and find the similarity using
1)Euclidean distance
2)cosine similarity


```python
import numpy as np
def euclid_dist(a,b):
  return np.linalg.norm(a-b)
def cosine_sim(a,b):
  dot_pdt=np.dot(a,b)
  norm_a=np.linalg.norm(a)
  norm_b=np.linalg.norm(b)
  cos_sim=dot_pdt/np.dot(norm_a,norm_b)
  return cos_sim

def main():
  a=list(map(int,input("Enter the elements of first vector:").split()))
  b=list(map(int,input("Enter the elements of second vector:").split()))
  a=np.array(a)
  b=np.array(b)
  euclidean_distance=euclid_dist(a,b)
  cosine_similarity=cosine_sim(a,b)
  print("Euclidean distance:",euclidean_distance)
  print("Cosine similarity:",cosine_similarity)

if __name__=="__main__":
  main()
```

```python
import pandas as pd
import os
import glob
train_folder="/content/drive/MyDrive/ML_LAB/extracted_data/ohsumed-first-20000-docs/training"
test_folder="/content/drive/MyDrive/ML_LAB/extracted_data/ohsumed-first-20000-docs/test"

def create_dfs(folder_path):
  dataframes=[]
  for root,dirs,files in os.walk(folder_path):
    for file in files:
      filename=os.path.join(root,file)
      with open(filename,'r') as f:
        row=f.read().split()
        df=pd.DataFrame([row])
        dataframes.append(df)
  if dataframes:
    return pd.concat(dataframes,ignore_index=True)
  else:
    return pd.DataFrame()

df1=create_dfs(train_folder)
df2=create_dfs(test_folder)
df1
df2
```
