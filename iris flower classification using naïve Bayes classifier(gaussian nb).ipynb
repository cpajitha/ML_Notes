Write a program to implement the naive bayes classifier for a sample training data set stored as a .CSV file.Compute the accuracy of the classifier,considering few datsets .

```python
from sklearn import datasets
iris=datasets.load_iris()
print("Features:",iris.feature_names)
print("Labels:",iris.target_names)
```

```python
x=iris.data
y=iris.target
import pandas as pd
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['target'] = iris.target
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn import naive_bayes
model=naive_bayes.GaussianNB()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
y_pred
```

```python
for i in y_pred:
  if i==0:
    print("setosa")
  elif i==1:
    print("versicolar")
  else:
    print("Virginica")
```

```python
y_test-y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
accuracy_in_percentage=accuracy_score(y_test,y_pred)*100
print("Accuracy = ",accuracy_in_percentage)
```
